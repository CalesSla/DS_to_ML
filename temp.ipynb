```python
import pandas as pd
import numpy as np

data = pd.read_csv("rent_apartments.csv")
data.head()
```

```python
data.dtypes
```

```python
data_encoded = pd.get_dummies(data, columns = ["balcony", "parking", "furnished", "garage", "storage"], drop_first=True, dtype=int)
data_encoded.head()
```

```python
data_encoded.garden.unique()
```

```python
import re

data_encoded.garden[4]
```

```python
int(re.findall(r'\d+', data_encoded.garden[4])[0])
```

```python
import warnings
warnings.filterwarnings("ignore")

for i in range(len(data_encoded)):
    if data_encoded.garden[i] == 'Not present':
        data_encoded.garden[i] = 0
    else:
        data_encoded.garden[i] = int(re.findall(f"\d+", data_encoded.garden[i])[0])
```

```python
data_encoded.head()
```

```python
X = data_encoded[["area", 
                  "constraction_year", 
                  "bedrooms", 
                  "garden",
                  "balcony_yes",
                  "parking_yes",
                  "furnished_yes",
                  "garage_yes",
                  "storage_yes"]]

y = data_encoded.rent
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

```python
rf.predict([[85, 2015,2 ,20, 1, 1, 0, 0, 1]])
```

```python
from sklearn.model_selection import GridSearchCV

grid_space = {"n_estimators": [100, 200, 300],
              "max_depth": [3, 6, 9, 12]}
grid = GridSearchCV(RandomForestRegressor(), param_grid=grid_space, cv=5, scoring='r2')

model_grid = grid.fit(X_train, y_train)

print(f"Best hyperparams are {model_grid.best_params_}, score = {model_grid.best_score_}")
```
